# Plotting a Line

```python
# importing necessary modules

import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
```

Sheriff travels 180 km/hr but is delayed by 5 minutes  
Equation: *y = 180(x - (1 / 12))*

```python
ys = 180 * (x - (1 / 12))
```

Robber travels 150 km/hr  
Equation: *y = 150x*

```python
yr = 150 * x
```

```python
figure, axes = plt.subplots()

# label the axes
plt.title("A Bank Robber Caught")
plt.xlabel("Time (in hours)")
plt.ylabel("Distance traveled (in kilometers)")

# set plot viewport
axes.set_xlim((0, 1))
axes.set_ylim((0, 100))

# plot the graph
axes.plot(x, ys, c="green")
axes.plot(x, yr, c="red")

# show intersection
plt.axvline(x=1/2, color="purple", linestyle="--")
plt.axhline(y=75, color="purple", linestyle="--")
```
